```python
### ваш код здесь ###

with open("_1984_.txt", "r", encoding="utf-8") as file:
    book_1984 = file.read()
print(book_1984[:100])
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')

nltk.download('punkt')
```

```python
### ваш код здесь ###
sentences_1984 = sent_tokenize(book_1984)
print(sentences_1984[:5])
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

Произведите разметку именованных сущностей для текста романа "1984", однако в этот раз не импортируйте `spacy`. Также не выводите ничего на экран!

Вместо этого сохраните результаты разметки в один большой список, а потом выведите на экран последние 5 элементов этого списка

```python
### ваш код здесь ###

ner_1984 = []

for sentence_1984 in sentences_1984:
    doc = nlp(sentence_1984)
    ner_1984.append([(ent.text, ent.label_) for ent in doc.ents])

print(ner_1984[:5])
```

```python
print(ner_1984[-1:-6])
```

```python
# Аннотирование PoS
for sentence_1984 in sentences_1984:
    doc = nlp(sentence_1984)
    print([(token.text, token.pos_) for token in doc])
```

Проделаем то же с текстом "1984" - не выводите ничего на экран, достаточно сохранить результаты в отдельный список. Выведите элементы в диапазоне от 10 до 20

```python
### ваш код здесь ###

pos_1984 = []

for sentence_1984 in sentences_1984:
    doc = nlp(sentence_1984)
    pos_1984.append([(token.text, token.pos_) for token in doc])
```

```python
print(pos_1984[10:20])
```

```python
from textblob import TextBlob
```

The sentiment property returns a namedtuple of the form Sentiment(polarity, subjectivity). The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
### ваш код здесь ###

polarity_1984 = []

for sentence_1984 in sentences_1984:
    blob = TextBlob(sentence_1984)
    polarity_1984.append(blob.sentiment.polarity)

print(f'Текст: {sentences_1984[4]}')
print(f'Тональность: {polarity_1984[4]}')
```

```python
data = {
      "sentence": sentences_1984,
      "ner": ner_1984,
      "pos": pos_1984,
      "sentiment": polarity_1984
  }
```
