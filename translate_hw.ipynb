# **Домашнее задание**

### 1. Выберите пару языков, для которой вы хотели бы создать систему машинного перевода на правилах


**русский язык - испанский язык**

Выбранная пара представляет интерес с точки зрения создания системы машинного перевода, поскольку указанные языки обладают значительными грамматическими различиями. Это проявляется на разных уровнях:
- письменность (разные алфавиты)
- падежная система (русский язык: шесть падежей, влияющих на форму некоторых частей речи; испанский язык: падежи отсутствуют, синтаксическая роль слова определяется его положением в предложении и предлогами)
- артикли (русский язык: артикли отсутствуют; испанский язык: определенные и неопределенные артикли, которые согласуются с существительными по роду и числу)
Также есть различия во временах глаголов, использовании личных местоимений и тп.



### 2. Как вы будете проводите пре-процессинг? Какие библиотеки вам понадобятся для пре-процессинга

Этапы пре-процессинга:

1.   Токенизация исходного текста
2.   Лемматизация
3.   Частеричная разметка
4.   Сопоставление синаксических структур

Необходимые библиотеки:

*   nltk (для токенизации)
*   spaСy (для токинизации, лемматизации испанского, морфологического разбора)
*   pymorphy2 (для лемматизации русского, морфологического разбора)
можно обойтись без pymorphy2, используя для лемматизации русского spaСy






### 3. Приведите 3 примера правил для этих языков

1.   Согласование по роду и числу

Необходимо отслеживать и сохранять согласование прилагательных и существительных как в испанском, так и в русском.

Для этого при помощи библиотек для морфологического разбора определяем в оригинале сначала существительное, потом согласованное с ним прилагательное. Определяем род и число. Находим соответствующие словам леммы в русском языке и ставим их в требуемую форму также при помощи библиотек.

2.   Склонение по падежам

При переводе с испанского на русский нужно определить падеж, который соответсвует русскому слову в переводе.

Для этого можно обратиться к заранее созданному словарю, где будут храниться глаголы с их управлением. Далее во время перевода алгоритм будет обращаться по лемме к словарю, определять управление и ставить нужное слово в трубемую форму.


3.   Артикли

Необходимо добавлять требуемый артикль при переводе с русского языка на испанский.

Для определения типа артикля (определенный / неопределенный / нулевой) можно воспользоваться поиском по близлежащим предложениям с целью понимания, в какой раз упоминается предмет (если в первый - неопределенный артикль, если не в первый - определенный). Далее необходимо при помощи библиотек для морфологического разбора определить род и число существительного и выбрать соответствующую форму для артикля в испанском.
